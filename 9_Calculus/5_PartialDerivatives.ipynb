```python
import sympy as sym
from IPython.display import display, Math
```

```python
x, y = sym.symbols('x, y')
```

```python
f = x**2 + x*y**2
```

```python
display(Math('\\frac {\\partial f}{\\partial x} = %s' %(sym.latex(sym.diff(f, x)))))
display(Math('\\frac {\\partial f}{\\partial y} = %s' %(sym.latex(sym.diff(f, y)))))
```

```python
p = sym.plotting.plot3d(f);
```

```python
p = sym.plotting.plot3d(sym.diff(f, x));
```

```python
p = sym.plotting.plot3d(sym.diff(f, y));
```
