```python
import sympy as sym
from IPython.display import display, Math
```

```python
x = sym.symbols('x')
```

```python
f = x
```

```python
sym.integrate(f)
```

```python
sym.integrate(f, (x, 0, 2))
```

```python
f = x**3 / (x - 2)
intf = sym.integrate(f)
p = sym.plot(f, show=False)
p.extend(sym.plot(intf, (x, 2.1, 10), show=False))
p[1].line_color = 'r'
p.ylim = [-200, 200]
p.xlim = [-10, 10]
p.show()
```
