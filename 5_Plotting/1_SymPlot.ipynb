```python
x, a = sym.symbols('x, a')
```

```python
expr = a / x
```

```python
sym.plot(expr.subs({a: 1}));
```

```python
p = sym.plot(expr.subs({a: 1}), show=False);
p.ylim = [-5, 5]
p[0].label = 'y = 1/x'
p.legend = True
p.show()
```

```python
p = sym.plot(expr.subs({a: 1}), show=False);
p.ylim = [-5, 5]
p[0].label = '$y = %s$' %(sym.latex(expr.subs({a: 1})))
p[0].line_color = 'r'

p2 = sym.plot(expr.subs({a: 2}), show=False);
p.extend(p2)
p[1].label = '$y = %s$' %(sym.latex(expr.subs({a: 2})))
p.legend = True
p.show()
```

```python
expr = a / (x**2 - a)
```

```python
p = sym.plot(expr.subs({a: 1}), show=False);
p.ylim = [-10, 10]
p.xlim = [-4, 4]
p[0].label = '$y = %s$' %(sym.latex(expr.subs({a: 1})))
p[0].line_color = 'r'

p2 = sym.plot(expr.subs({a: 2}), show=False);
p.extend(p2)
p[1].label = '$y = %s$' %(sym.latex(expr.subs({a: 2})))
p.legend = True
p.show()
```
