```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Polygon
```

```python
y = np.array([[1,1], [2,3], [3,1]])
p = Polygon(y)
p
```

```python
fig, ax = plt.subplots()
ax.add_patch(p)
ax.set_xlim([0, 4])
ax.set_ylim([0, 4])
plt.show()
```

```python
p = Polygon(y, facecolor='m', alpha=0.3)
fig, ax = plt.subplots()
ax.add_patch(p)
ax.set_xlim([0, 4])
ax.set_ylim([0, 4])
plt.show()
```

```python
p = Polygon(y, facecolor='m', alpha=0.3)
y1 = np.array([[2,2], [2.5,4], [3.5,1]])
p1 = Polygon(y1, facecolor='r', alpha=0.3, edgecolor='k')
fig, ax = plt.subplots()
ax.add_patch(p)
ax.add_patch(p1)
ax.set_xlim([0, 4])
ax.set_ylim([0, 4])
plt.show()
```

```python
x = np.linspace(-2, 2, 101)
f = -x**2
```

```python
plt.plot(x, f);
```

```python
pnts = np.vstack((x, f))
pnts.shape
```

```python
pnts = pnts.T
pnts.shape
```

```python
p = Polygon(pnts, facecolor = 'g', edgecolor = 'k', alpha=0.2)
fig, ax = plt.subplots()
ax.add_patch(p)
ax.set_xlim([-2, 2])
ax.set_ylim([-4, 0])
plt.plot(x, f, 'k')
plt.show()
```
