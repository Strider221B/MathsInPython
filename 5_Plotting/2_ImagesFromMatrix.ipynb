```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.zeros((10, 14))
for i in range(10):
    for j in range(14):
        A[i, j] = 3*i - 4*j
```

```python
plt.imshow(A);
```

```python
plt.figure(figsize=(20, 10))
plt.imshow(A)
for i in range(10):
    for j in range(14):
        plt.text(j, i, A[i, j])
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.imshow(A)
for i in range(10):
    for j in range(14):
        plt.text(j, i, A[i, j], horizontalalignment='center')
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.imshow(A)
for i in range(10):
    for j in range(14):
        plt.text(j, i, A[i, j], horizontalalignment='center')
plt.set_cmap('Purples')
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.imshow(A)
for i in range(10):
    for j in range(14):
        plt.text(j, i, A[i, j], horizontalalignment='center')
plt.set_cmap('twilight_shifted')
plt.show()
```
