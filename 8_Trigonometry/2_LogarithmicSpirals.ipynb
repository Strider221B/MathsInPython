### Formula

x = cos(t)$e^{t/k}$

y = sin(t)$e^{t/k}$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 10*np.pi, 1234)
k = -3
```

```python
x = np.cos(t) * np.exp(t / k)
y = np.sin(t) * np.exp(t / k)
```

```python
plt.plot(x, y);
```

```python
N = 150
ks = np.linspace(-6, -2, N)
```

```python
for i in range(N):
    k = ks[i]
    x = np.cos(t) * np.exp(t / k)
    y = np.sin(t) * np.exp(t / k)
    plt.plot(x, y)

plt.axis('off')
plt.show()
```

```python
for i in range(N):
    k = ks[i]
    x = np.cos(t) * np.exp(t / k)
    y = np.sin(t) * np.exp(t / k)
    plt.plot(x, y, color=[i/N, 0, i/N])

plt.axis('off')
plt.show()
```
