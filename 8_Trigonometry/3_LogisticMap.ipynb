### Formula

$x_{n+1}$ = $r{x_n} (1 - {x_n})$

Has utilization in Chaos, Biology, Physics and Mathematics.

#### Here we will use:
${x_1}$ = 0.5

$n \in N : n <= 500$

$r \in$ [1, 4] in 1000 steps

For each value of $r$, plot the unique values in the last 10% of the vector $x$.

X Axis: $r$ & Y Axis: $x$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 500
x = np.zeros(N)
x[0] = 0.5
```

```python
%matplotlib notebook
# plt.figure(figsize=(15, 10))
for r in range(5):
    for n in range(N-1):
        x[n+1] = r*x[n]*(1 - x[n])
    plt.plot(x, 's-', label=r);

plt.legend()
plt.show()
```

```python
N = 500
x = np.zeros(N)
x[0] = 0.5
rs = np.linspace(1, 4, 1000)
end_indices = np.arange(int(N*0.9), N)
```

```python
for r in rs:
    for n in range(N-1):
        x[n+1] = r*x[n]*(1 - x[n])
    end_values = x[end_indices]
    end_values = np.unique(end_values)
    r_for_this = np.ones(end_values.shape) * r
    plt.plot(r_for_this, end_values, 'k.', markersize=0.1);
plt.show()
```

```python
for r in rs:
    for n in range(N-1):
        x[n+1] = r*x[n]*(1 - x[n]) + 0.2 * np.sin(2 * np.pi * x[n]) ** 2
    end_values = x[end_indices]
    end_values = np.unique(end_values)
    r_for_this = np.ones(end_values.shape) * r
    plt.plot(r_for_this, end_values, 'k.', markersize=0.1);
plt.show()
```
