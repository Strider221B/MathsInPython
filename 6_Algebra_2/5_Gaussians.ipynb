```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
h = 0.999
t = np.linspace(-3, 3, 1001)
g = np.exp(-4*np.log(2)*t**2 / h**2)
```

```python
plt.plot(t, g);
```

```python
N = 100
G = np.zeros((N, len(t)))
```

```python
for i in range(N):
    h = (i + 1) / N
    G[i, :] = np.exp(-4*np.log(2)*t**2 / h**2)
```

```python
plt.figure(figsize=(5, 5))
plt.imshow(G);
```

```python
plt.figure(figsize=(5, 5))
plt.pcolormesh(G);
```

```python
h = np.zeros(N)
for i in range(N):
    h[i] = (i + 1) / N
    G[i, :] = np.exp(-4*np.log(2)*t**2 / h[i]**2)
```

```python
plt.figure(figsize=(5, 5))
plt.pcolormesh(t, h, G);
```
