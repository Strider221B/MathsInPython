Even function: f(-x) = f(x)

Odd function: f(-x) = -f(x)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-5, 5, 20)
```

```python
f_even = x**2
f_even_neg = (-x)**2
```

```python
plt.plot(x, f_even)
plt.plot(x, f_even_neg, 'ro');
```

```python
f_odd = x**3
f_odd_neg = (-x)**3
f_neg_odd = -f_odd
```

```python
plt.plot(x, f_odd)
plt.plot(x, f_odd_neg, 'ro')
plt.plot(x, f_neg_odd, 'r');
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 39)
```

```python
cos_fun = np.cos(x)
cos_fun_n = np.cos(-x)

plt.plot(x, cos_fun)
plt.plot(x, cos_fun_n, 'ro');
```

```python
sin_fun = np.sin(x)
sin_fun_n = np.sin(-x)

plt.plot(x, sin_fun)
plt.plot(x, sin_fun_n, 'ro');
```
