```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as sps
```

```python
x = np.linspace(0, 12*np.pi, 213)
fx = -np.cos(x) - np.sqrt(x)
plt.plot(x, fx);
```

```python
peaks = sps.find_peaks(fx)
peaks
```

```python
peaks[0]
```

```python
plt.plot(x, fx)
plt.plot(x[peaks[0]], fx[peaks[0]], 'o');
```
