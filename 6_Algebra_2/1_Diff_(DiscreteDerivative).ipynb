```python
x = np.linspace(-2, 2, 100)
fx = x**2
```

```python
dfx = np.diff(fx)
```

```python
plt.plot(x, fx)
plt.plot(x[:-1], dfx)
plt.show()
```

```python
dfx_dx = np.diff(fx) / np.diff(x)
```

```python
plt.plot(x, fx)
plt.plot(x[:-1], dfx_dx)
plt.ylim(-2, 4)
plt.show()
```
