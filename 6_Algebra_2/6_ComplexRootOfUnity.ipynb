$z^n = 1$ z $\in$ complex

$z = e^{2\pi ik / n}$ ; k = 1, 2, 3, ..., n-1

```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
from IPython.display import display, Math
```

```python
n = 5
for k in range(n):
    z = np.exp(2 * np.pi * 1j * k / n)
    print(z**n)
```

```python
n = 5
for k in range(n):
    z = np.exp(2 * np.pi * 1j * k )
    print(z)
```

```python
n = 5
for k in range(n):
    z = sym.exp(2 * sym.pi * sym.I * k )
    print(z)
```

```python
n = 5
for k in range(n):
    z = sym.exp(2 * sym.pi * sym.I * k / n )
    display(Math('(%s)^{%s} \\Rightarrow %s' %(sym.latex(z), n, sym.latex(z**n))))
```

```python
n = 4
for k in range(n):
    z = sym.exp(2 * sym.pi * sym.I * k / n )
    display(Math('(%s)^{%s} \\Rightarrow %s' %(sym.latex(z), n, sym.latex(z**n))))
```

```python
n = 125
for k in range(n):
    z = np.exp(2 * np.pi * 1j * k / n )
    plt.plot([0, z.real], [0, z.imag])
plt.axis('square');
```

```python
n = 125
for k in range(n):
    z = k * np.exp(2 * np.pi * 1j * k / n )
    plt.plot([0, z.real], [0, z.imag])
plt.axis('off');
```

```python
n = 125
for k in range(n):
    z = k * np.exp(2 * np.pi * 1j * k / n )
    plt.plot([0, z.real], [0, z.imag], color=[k/n, k/n, k/n])
plt.axis('off');
```
