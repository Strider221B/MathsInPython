```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
n = 1000

x = np.random.randn(n)
y = np.random.randn(n)**2
```

```python
plt.hist2d(x, y, 30, vmax=10);
```

```python
sns.jointplot(x=x, y=y);
```

```python
sns.set()
sns.jointplot(x=x, y=y);
```

```python
sns.jointplot(x=x, y=y, kind='reg');
```

```python
sns.jointplot(x=x, y=y, kind='kde');
```

```python
sns.jointplot(x=x, y=y).plot_joint(sns.kdeplot);
```
