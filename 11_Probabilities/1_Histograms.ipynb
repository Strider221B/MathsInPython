#### Plot distribution for:

$ y = x; y = x^2; y = x^3$

```python
x = np.linspace(-3, 3, 10000)
```

```python
_, axs = plt.subplots(2, 3, figsize=(10, 6))
for i in range(3):
    y = x ** (i+1)
    h = np.histogram(y, 50)
    yP = h[0] / np.sum(h[0])
    
    ax = axs[0, i]
    ax.plot(x, y)
    ax.set_title('$y = x^{%g}$' %(i+1))
    
    ax = axs[1, i]
    ax.plot(h[1][:-1], yP)
    ax.set_title('Probability Density')
plt.show()
```
